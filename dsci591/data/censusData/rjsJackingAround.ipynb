```python
def importBz2( file ):
    start = time.time()
    data = bz2.BZ2File(file,'rb')
    # newDF = pd.read_pickle('allTogetherNow.pkl')
    return pd.read_csv(data), print(time.time()-start)

newDF = importBz2('allTogetherNow.bz2')
```

```python
def importPbz2( file ):
    start = time.time()
    data = bz2.BZ2File(file,'rb')
    # newDF = pd.read_pickle('allTogetherNow.pkl')
    return pd.read_pickle(data), print(time.time()-start)

newDF = importPbz2('allTogetherNow.pbz2')
```

```python
def covidPollutantFipsCorr(df, pollutant, cols=[]):
    if cols:
        polDF =  df.loc[df['pollutant']==pollutant,cols+['fips']]
    else:
        polDF = df.loc[df['pollutant']==pollutant,:]
    fipsDF  = polDF.groupby(by='fips').agg('mean')
    return fipsDF.corr()
```
